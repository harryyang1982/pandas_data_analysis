## 1. 함수 매핑
### 1-1 개별 원소에 함수 매핑

```python
import seaborn as sns

titanic = sns.load_dataset('titanic')
df = titanic.loc[:, ['age', 'fare']]
df['ten'] = 10
df.head()
```

```python
def add_10(n):
    return n + 10

def add_two_obj(a, b):
    return a + b

print(add_10(10))
print(add_two_obj(10, 10))
```

```python
sr1 = df['age'].apply(add_10)
sr1.head()
```

```python
sr2 = df['age'].apply(add_two_obj, b = 10)
sr2.head()
```

```python
sr3 = df['age'].apply(lambda x: add_10(x)) # x=df['age']
sr3.head()
```

```python
df_map = df.applymap(add_10) # 모든 원소에 10 더하기
df_map.head()
```

```python
df
```

```python
df_map2 = df.applymap(add_two_obj, b = 500)
df_map2.head()
```

### 1-2 시리즈 객체에 함수 매핑

```python
def missing_value(series):
    return series.isnull()

# 데이터프레임에 apply 메소드를 적용
result = df.apply(missing_value, axis=0)
result.head()
```

```python
type(result)
```

```python
def min_max(x):
    return x.max() - x.min()

result = df.apply(min_max, axis=0)
result
```

```python
type(result)
```

```python
df['add'] = df.apply(lambda x: add_two_obj(x['age'], x['ten']), axis = 1)
df.head()
```

```python
df['add2'] = df['age'] + df['ten']
```

```python
df
```

### 1-3 데이터프레임 객체에 함수 매핑

```python
def missing_value(x):
    return x.isnull()

def missing_count(x):
    return missing_value(x).sum()

def total_number_missing(x):
    return missing_count(x).sum()
```

```python
result_df = df.pipe(missing_value)
result_df.head()
```

```python
result_series = df.pipe(missing_count)
result_series
```

```python
type(result_series)
```

```python
result_value = df.pipe(total_number_missing)
result_value
```

```python
type(result_value)
```

## 3. 필터링
### 3-1 불린 인덱싱

```python
import seaborn as sns
titanic = sns.load_dataset('titanic')

mask1 = (titanic.age >= 10) & (titanic.age < 20)
df_teenage = titanic.loc[mask1, :]
df_teenage.head()
```

```python
mask2 = (titanic.age < 10) & (titanic.sex == 'female')
df_female_under10 = titanic.loc[mask2, :]
df_female_under10.head()
```

```python
mask3 = (titanic.age < 10) | (titanic.age >= 60)
df_under10_morethan60 = titanic.loc[mask3, ['age', 'sex', 'alone']]
df_under10_morethan60.head()
```

### 3-2 isin() 메소드 활용

```python
import seaborn as sns
import pandas as pd

titanic = sns.load_dataset('titanic')

pd.set_option('display.max_columns', 10)

mask3 = titanic['sibsp'] == 3
mask4 = titanic['sibsp'] == 4
mask5 = titanic['sibsp'] == 5

df_boolean = titanic[mask3 | mask4 | mask5]
df_boolean.head()
```

```python
isin_filter = titanic['sibsp'].isin([3, 4, 5]) # %in% c(3, 4, 5) in R
df_isin = titanic[isin_filter]
df_isin.head()
```

## 4. 데이터프레임 합치기
### 4-1 데이터프레임 연결

```python
import pandas as pd

df1 = pd.DataFrame({'a': ['a0', 'a1', 'a2', 'a3'],
                    'b': ['b0', 'b1', 'b2', 'b3'],
                    'c': ['c0', 'c1', 'c2', 'c3']},
                   index=[0, 1, 2, 3])

df2 = pd.DataFrame({'a': ['a2', 'a3', 'a4', 'a5'],
                    'b': ['b2', 'b3', 'b4', 'b5'],
                    'c': ['d2', 'd3', 'd4', 'd5']},
                   index=[2, 3, 4, 5])

print(df1, '\n')
print(df2, '\n')
```

```python
result1 = pd.concat([df1, df2])
result1
```

```python
result2 = pd.concat([df1, df2], ignore_index=True)
result2
```

```python
result3 = pd.concat([df1, df2], axis=1)
result3
```

```python
result3_in = pd.concat([df1, df2], axis=1, join = 'inner')
result3_in
```

```python
sr1 = pd.Series(['e0', 'e1', 'e2', 'e3'], name='e')
sr2 = pd.Series(['f0', 'f1', 'f2'], name='f', index=[3, 4, 5])
sr3 = pd.Series(['g0', 'g1', 'g2', 'g3'], name='g')
result4 = pd.concat([df1, sr1], axis=1)
```

```python
result4
```

```python
result5 = pd.concat([df2, sr2], axis=1, sort=True)
result5
```

```python
result6 = pd.concat([sr1, sr3], axis=1)
print(result6, '\n')

result7 = pd.concat([sr1, sr3], axis=0)
print(result7)
```
