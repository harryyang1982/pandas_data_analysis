```python
import time
import random
import pandas as pd
from selenium import webdriver

driver = webdriver.Chrome(r"/Users/seunghoonyang/Documents/python/chromedriver")
driver.get("http://www.statiz.co.kr/player.php?opt=1&name=%EC%96%91%EC%A4%80%ED%98%81&birth=1969-05-26")

df = pd.DataFrame()

for n in range(3, 20) : # len(driver.find_elements_by_css_selector('tbody > tr'))
    info = driver.find_element_by_xpath(f"/html/body/div[1]/div[1]/div/section[2]/div/div[2]/div/div[3]/div/div/table/tbody/tr[{n}]").text.split(" ")
    df = df.append(pd.DataFrame(info).T)

df.to_csv("./info_df.txt", sep = "|", encoding = 'utf-8')
```
