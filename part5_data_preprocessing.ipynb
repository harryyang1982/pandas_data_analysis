## 4. 범주형(카테고리) 데이터 처리

### 4-1 구간 분할

```python
import pandas as pd
import numpy as np
df = pd.read_csv("./data/part5/auto-mpg.csv", header=None,
                 names=['mpg', 'cylinders', 'displacement', 'horsepower', 'weight',
              'acceleration', 'model year', 'origin', 'name'])

df['horsepower'].replace('?', np.nan, inplace=True)
df.dropna(subset=['horsepower'], axis=0, inplace=True)
df['horsepower']=df['horsepower'].astype('float')
```

```python
count, bin_dividers = np.histogram(df['horsepower'], bins=3)
bin_dividers
```

```python
bin_names = ['저출력', '보통출력', '고출력']

df['hp_bin'] = pd.cut(x=df['horsepower'],
                      bins=bin_dividers,
                      labels=bin_names,
                      include_lowest=True) # 첫 경계값 포함

df[['horsepower', 'hp_bin']].head(15)
```

### 4-2 더미 변수

```python
count, bin_dividers = np.histogram(df['horsepower'], bins=3)

bin_names = ['저출력', '보통출력', '고출력']

df['hp_bin'] = pd.cut(x=df['horsepower'],
                      bins=bin_dividers,
                      labels=bin_names,
                      include_lowest=True) # 첫 경계값 포함

horsepower_dummies = pd.get_dummies(df['hp_bin'])
```

```python
# 전처리를 위한 encoder 객체 만들기

label_encoder = preprocessing.LabelEncoder()
onehot_encoder = preprocessing.OneHotEncoder()

```

## 5. 정규화

```python
import pandas as pd
import numpy as np

df = pd.read_csv("./data/part5/auto-mpg.csv", header=None,
                 names=['mpg', 'cylinders', 'displacement', 'horsepower', 'weight',
              'acceleration', 'model year', 'origin', 'name'])

df['horsepower'].replace('?', np.nan, inplace=True)
df.dropna(subset=['horsepower'], axis=0, inplace=True)
df['horsepower']=df['horsepower'].astype('float')
```

```python
df.horsepower.describe()
```

```python
df["std_horsepower"] = df.horsepower/abs(df.horsepower.max())
```

```python
min_x = df.horsepower - df.horsepower.min()
min_max = df.horsepower.max() - df.horsepower.min()
df["range_horsepower"] = min_x/min_max
```

**Timestamp to Period**

```python
import pandas as pd

dates = ['2019-01-01', '2020-03-01', '2021-06-01']
ts_dates = pd.to_datetime(dates)
print(ts_dates)
```

```python
# timestamp to period

pr_day = ts_dates.to_period(freq='D')
print(pr_day)
pr_month = ts_dates.to_period(freq='M')
print(pr_month)
pr_year = ts_dates.to_period(freq='Y')
print(pr_year)
```

### 6-2 시계열 데이터 만들기

```python
import pandas as pd

ts_ms = pd.date_range(start='2019-01-01',
                      end=None,
                      periods=6,
                      freq='MS',
                      tz='Asia/Seoul')
print(ts_ms)
```

```python
ts_me = pd.date_range('2019-01-01', periods=6,
                      freq='M',
                      tz='Asia/Seoul')

print(ts_me)

ts_3m = pd.date_range('2019-01-01', periods=6,
                      freq='3M',
                      tz='Asia/Seoul')

print(ts_3m)
```

```python
pr_m = pd.period_range(start='2019-01-01', end=None,
                       periods=3, freq='M')

print(pr_m)
```

```python
pr_h = pd.period_range(start='2019-01-01',
                       end=None,
                       periods=3,
                       freq='H')

print(pr_h)
```
